```r
# carrega a package readr
# The goal of readr is to provide a fast and friendly way to read rectangular data 
# (like csv, tsv, and fwf). It is designed to flexibly parse many types of data found in the wild, 
# while still cleanly failing when data unexpectedly changes.
library (readr)
```

```r
# localizacao do dataset no portal github
# https://raw.githubusercontent.com/HeronCarlos/EDS-MOD01/master/netflix_titles.csv
```

```r
urldataset = "https://raw.githubusercontent.com/HeronCarlos/EDS-MOD01/master/netflix_titles.csv"
```

```r
netflix <- read_csv(url(urldataset))
```

```r
# visualizacao dos dados 
head(netflix)
```

```r
# aplica a library DBI - para conexao em databases utilizando protocolo ODBC
library(DBI)
```

```r
con <- dbConnect(odbc::odbc(), "PostgreSQL35W", timeout = 10)
con
df <- dbReadTable(con,"Test2")

print(df)
```

```r
# cria uma datafame do nosso dataset netflix
dfnetflix = as.data.frame(netflix)
```

```r
head(dfnetflix)
```

```r
# efetua a escrita em banco de dados Postgresql atraves da conexao ODBC ,com configuracao para 
# o banco de dados EDS
# cria a tabela 
dbWriteTable(con, "netflix_titles", dfnetflix)
# caso for reutilizar , set o parametro overwrite = true
# dbWriteTable(con, "netflix_titles", dfnetflix, overwrite = TRUE)
# caso for reutilizar, incluir linhas , set o parametro append = true
# dbWriteTable(con, "netflix_titles", dfnetflix, append = TRUE)
```
