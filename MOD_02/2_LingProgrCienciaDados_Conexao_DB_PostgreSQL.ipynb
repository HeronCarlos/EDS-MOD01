```python
# IMPORT LIBRARY
import pandas as pd 
import psycopg2 as pg 
```

```python
# definindo uma conexao 
# dbname – the database name (database is a deprecated alias)
# user – user name used to authenticate
# password – password used to authenticate
# host – database host address (defaults to UNIX socket if not provided)
# port – connection port number (defaults to 5432 if not provided)

conn = pg.connect(dbname = "dvdrental", user="postgres", password="carpediem", host="localhost")
```

```python
# cria um cursor para percorrer o conjunto de dados 
cur = conn.cursor()
```

```python
cur.execute("SELECT film_id, title, release_year FROM public.film;")
```

```python
records = cur.fetchall()
```

```python
# todas as linhas
records
```

```python
# Imprime elementos especificos de um registro
for row in records:
    print('film_id: {0} title: {1}'.format(row[0], row[1]))
```

```python
# criacao de um dataframe pandas com o resultado do select acima
dfregistros = pd.DataFrame(records)
```

```python
# imprime as 5 primeiras linhas do dataframe
dfregistros.head()
```

```python
#fecha o cursor
cur.close()
```

```python
# fecha a conexao com o banco de dados
conn.close()
```
